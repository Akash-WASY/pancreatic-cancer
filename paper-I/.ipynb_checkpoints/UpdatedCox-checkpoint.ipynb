```python
# Import required libraries :
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# import pandas_profiling
from lifelines import KaplanMeierFitter
```

```python
mtdata = pd.read_excel("Mean Methylation Univariate freash.xlsx", sheet_name="KNN")
```

```python
mtdata
```

```python
T = mtdata["CLI_days_to_death"]
E = mtdata["CLI_Time"]
plt.hist(T, bins = 50)
plt.xlabel("Histogram of days_to_death variable")
plt.show()
```

```python
kmf_s = KaplanMeierFitter()
kmf_s.fit(mtdata["CLI_Time"], mtdata["CLI_vital_status"])

kmf_s.survival_function_
```

```python
kmf_s.plot_survival_function()
```

```python
mtdata.drop(['ProbeID', 'tcga_participant_barcode','CLI_days_to_last_followup'], axis=1)
from lifelines import CoxPHFitter
rossi_dataset = mtdata.drop(['ProbeID', 'tcga_participant_barcode','CLI_days_to_last_followup', 'CLI_days_to_death'], axis=1)
cph = CoxPHFitter()
cph.fit(rossi_dataset, duration_col='CLI_Time', event_col='CLI_vital_status')
cph.print_summary() 
```

```python
cph.plot()
```

```python
cph.predict_survival_function(mtdata).plot(legend=None)
```

```python
cph_data = cph.summary.query('p < 0.05')
cph_data
```

```python
np.mean(cph.baseline_hazard_['hazard'])
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_nomogram(data, predictors, coefficients, baseline_hazard):
    plt.figure(figsize=(10, 6))
    for i in range(len(predictors)):
        plt.plot([i, i + 1], [0, coefficients[i]], 'k-', lw=2)
        plt.text(i + 0.5, coefficients[i] / 2, predictors[i], ha='center', va='center')
    plt.plot([0, len(predictors)], [0, 0], 'k-', lw=2)
    plt.plot([0, 0], [0, np.exp(baseline_hazard)], 'k-', lw=2)
    plt.text(0, np.exp(baseline_hazard) / 2, 'Baseline hazard', ha='center', va='center')
    plt.xlim(-1, len(predictors) + 1)
    plt.ylim(0, np.exp(np.max(coefficients)))
    plt.gca().set_yticks([np.exp(x) for x in np.arange(np.min(coefficients), np.max(coefficients), 2)])
    plt.gca().set_yticklabels(['{:.2f}'.format(np.exp(x)) for x in np.arange(np.min(coefficients), np.max(coefficients), 2)])
    plt.xticks(range(len(predictors) + 1), [''] + predictors + [''])
    plt.xlabel('Predictors')
    plt.ylabel('Hazard Ratio')
    plt.title('Nomogram for Cox Regression Analysis')
    plt.show()

predictors = ['Age', 'Gender', 'BMI', 'Smoker']
coefficients = [-0.15, 0.25, 0.10, 0.35]
baseline_hazard = -0.75

plot_nomogram(mtdata, predictors, coefficients, baseline_hazard)
```
